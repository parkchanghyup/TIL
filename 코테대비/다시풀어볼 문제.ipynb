```python
# 가장 흔한 단어
import collections
import re
def mostCommonWord( paragraph: str, banned: List[str]) -> str:
    
    words = [x.lower() for x  in re.sub(r'[^\w]',' ',paragraph).split() 
             if x not in banned ]
    
    answer  =  collections.Counter(words)
    return answer.most_common(1)[0][0]
    
paragraph = "Bob hit a ball, the hit BALL flew far after it was hit."
banned = ["hit"]

mostCommonWord(paragraph,banned)
```

```python
# 두수의 합
'''
Input: nums = [2,7,11,15], target = 9
Output: [0,1]
'''

from typing import List
import collections

def twoSum( nums: List[int], target: int) -> List[int]:
    dic = collections.defaultdict(int)
    
    for idx,num in enumerate(nums):
        dic[num]=idx
        
    for idx,num in enumerate(nums):
        if  target - num in dic and idx !=dic[target-num]:
            return nums.index(num), dic[target-num]
        

twoSum ( [3,3],6) 
        
```

```python
# 세수의 합
'''
Input: nums = [-1,0,1,2,-1,-4]
Output: [[-1,-1,2],[-1,0,1]]
'''


def threeSum(nums: List[int]) -> List[List[int]]:

    results = []
    nums.sort()

    for i in range(len(nums) - 2):
        # 중복된 값 건너뛰기
        if i > 0 and nums[i] == nums[i - 1]:
            continue

        #간격을 좁혀 가며 sum 계산
        left, right = i + 1, len(nums) - 1
        while left < right:
            sum_num = nums[i] + nums[left] + nums[right]
            if sum_num < 0:
                left += 1
            elif sum_num > 0:
                right -= 1
            else:
                # sum = 0인 경우이므로 정답 및 스킵 처리
                results.append([nums[i], nums[left], nums[right]])

                while left < right and nums[left] == nums[left + 1]:
                    left += 1
                while left < right and nums[right] == nums[right - 1]:
                    right -= 1
                left += 1
                right -= 1
                
    return results
```

```python
# 중복 문자 제거
# https://leetcode.com/problems/remove-duplicate-letters/submissions/
'''
Input: s = "bcabc"
Output: "abc"
'''
import collections

def removeDuplicateLetters(s: str) -> str:
    counter,seen,stack = collections.Counter(s),set(),[]
        
    for char in s :
        counter[char] -=1
        if char in seen :
            continue
        # 뒤에 붙일 문자가 남아 있다면 스택에서 제거
        while stack and char < stack [-1] and counter[stack[-1]]  > 0:
            seen.remove(stack.pop())
        stack.append(char)
        seen.add(char)
        print(stack)
    
    return (''.join(stack))
removeDuplicateLetters ('cbacdcbc')       
```

```python
# 조합의 합
from typing import List

def combinationSum(candidates: List[int], target: int) -> List[List[int]]:
    result = []

    def dfs(csum, index, path):
        # 종료 조건
        if csum < 0:
            return
        elif csum == 0:
            result.append(path)
            return
        
        #자신부터 하위 원소까지의 나열  재귀 호출
        for i in range(index, len(candidates)):
            dfs(csum -candidates[i], i, path+[candidates[i]])

    dfs(target, 0, [])
    return result


combinationSum([2, 3, 6, 7], 7)
```

```python
import heapq
from typing import List
import collections
# k경유지 내 가장 저렴한 항공권
'''
Input: 
n = 3, edges = [[0,1,100],[1,2,100],[0,2,500]]
src = 0, dst = 2, k = 1
Output: 200
'''
import heapq


def findCheapestPrice(n: int, edges: List[List[int]], src: int, dst: int,
                      K: int) -> int:
    graph = collections.defaultdict(list)

    # 그래프 생성
    for u, v, w in edges:
        graph[u].append((v, w))

    Q = [(0, src, K)]

    while Q:
        time, node, k = heapq.heappop(Q)

        if node == dst:
            return time
        if k >= 0:
            for v, w in graph[node]:
                heapq.heappush(Q, (w + time, v, k - 1))

    return -1


findCheapestPrice(n=3,
                  edges=[[0, 1, 100], [1, 2, 100], [0, 2, 500]],
                  src=0,
                  dst=2,
                  K=1)
```

```python
# 주유소
'''
Input: 
gas  = [1,2,3,4,5]
cost = [3,4,5,1,2]
'''


def canCompleteCircuit( gas: List[int], cost: List[int]) -> int:
    if sum(gas) < sum(cost):
        return -1
    
    start,fuel = 0,0
    
    for i in range(len(gas)):
        
        # 출발점이 안되는 지점 판별
        if gas[i] + fuel< cost[i]:
            start = i+1
            fuel = 0
        else :
            fuel +=gas[i]-cost[i]
    return start
        
```

```python
# 괄호를 삽입하는 여러 가지 방법

def diffWaysToCompute(self, input: str) -> List[int]:
    
    if input.isdigit():
        return [int(input)]
    
    def compute(left, right, op):
        results = []
        for l in left:
            for r in right:
                results.append(eval(str(l)+op+str(r)))
        return results:
        
    results = []
    
    for index, value in enumerate(input):
        if value in '-+*':
            left = self.diffWaysToCompute(input[:index])
            right = self.diffWaysToCompute(input[index+1:])

            results.extend(compute(left, right, value))

    return results
```

```python
# 가사 검색

import bisect
import collections

def func(a,left,right):
    left_idx = bisect.bisect_left(a,left)
    right_idx = bisect.bisect_right(a,right)
    
    return right_idx - left_idx

def solution(words, queries):
    answer = []
    
    dic = collections.defaultdict(list)
    dic_reverse = collections.defaultdict(list)
    for word in words:
        # 단어 길이 순으로 분리
        dic[len(word)].append(word)
        dic_reverse[len(word)].append(word[::-1])
    
    
    #정렬
    for key in dic.keys():
        dic[key].sort()
        dic_reverse[key].sort()
    
    for query in queries: # 쿼리를 하나씩 확인하며 처리
        #접미사에 와일드 카드가 붙은 경우
        if query[0] != '?':
            answer.append(func(dic[len(query)],query.replace('?','a'),query.replace('?','z')))
            
        #접두사에 와일드 카드가 붙은 경우
        else :
            query = query[::-1]
            answer.append(func(dic_reverse[len(query)],query.replace('?','a'),query.replace('?','z')))
    return answer


solution(["frodo", "front", "frost", "frozen", "frame", "kakao"], ["fro??", "????o", "fr???", "fro???", "pro?"])
```

```python
# 행성 터널

N = int(input())

points=[]
for i in range(N):
    point = map(int,input().split())
    points.append(point)
    

def find_parent(parent, x):
    if parent[x] != x:
        parent[x] = find_parent(parent, parent[x])
    return parent[x]


def union_parent(parent, a, b):
    a = find_parent(parent, a)
    b = find_parent(parent, b)

    if a < b:
        parent[b] = a
    else:
        parent[a] = b


parent = [0] * (N+1)

for i in range(1, N+1):
    parent[i] = i

x = []
y = []
z = []
i = 1
for a, b, c, in points:

    x.append((a, i))
    y.append((b, i))
    z.append((c, i))
    i += 1

x.sort()
y.sort()
z.sort()

edges = []

for i in range(N-1):
    edges.append((x[i+1][0] - x[i][0], x[i+1][1], x[i][1]))
    edges.append((y[i+1][0] - y[i][0], y[i+1][1], y[i][1]))
    edges.append((z[i+1][0] - z[i][0], z[i+1][1], z[i][1]))

edges.sort()

result = 0
for cost,a,b in edges:
    
    if find_parent(parent, a) == find_parent(parent,b):
        continue
    else:
        union_parent(parent,a,b)
        result += cost

print(result)
```

```python
def solution(n, delivery):

    # 0은 모름 1은 재고 o -1은 재고 x

    answer = [0] * (n + 1)

    new_check_list = []

    # 재고 있는것들 처리
    for a, b, pos in delivery:
        if pos == 1:
            answer[a] = 1
            answer[b] = 1
        else:
            new_check_list.append([a, b, pos])

    # 재고가 확실히 없는 것들 처리
    for a, b, pos in new_check_list:

        if answer[a] == 1:
            answer[b] = -1
        elif answer[b] == 1:
            answer[a] = -1

    result = ''

    for i in range(1, n + 1):
        if answer[i] == 0:
            result += '?'
        elif answer[i] == 1:
            result += 'O'
        else:
            result += 'X'
    return result
```
