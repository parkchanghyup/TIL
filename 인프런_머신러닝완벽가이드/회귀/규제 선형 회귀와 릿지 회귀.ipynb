```python
#  LinearRegression  feature   X_data Target   Y_target    
from sklearn.linear_model import Ridge
from sklearn.model_selection import cross_val_score
import pandas as pd
import numpy as np
import seaborn as sns
from scipy import stats
from sklearn.datasets import load_boston
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error , r2_score
%matplotlib inline

# boston  
boston = load_boston()

# boston  DataFrame  
bostonDF = pd.DataFrame(boston.data , columns = boston.feature_names)

# boston dataset target array  .  PRICE  DataFrame . 
bostonDF['PRICE'] = boston.target
print('Boston   :',bostonDF.shape)

y_target = bostonDF['PRICE']
X_data = bostonDF.drop(['PRICE'],axis=1,inplace=False)

X_train , X_test , y_train , y_test = train_test_split(X_data , y_target ,test_size=0.3, random_state=156)

ridge = Ridge(alpha = 10)
neg_mse_scores = cross_val_score(ridge, X_data, y_target, scoring="neg_mean_squared_error", cv = 5)
rmse_scores  = np.sqrt(-1 * neg_mse_scores)
avg_rmse = np.mean(rmse_scores)
print(' 5 folds   Negative MSE scores: ', np.round(neg_mse_scores, 3))
print(' 5 folds   RMSE scores : ', np.round(rmse_scores,3))
print(' 5 folds   RMSE : {0:.3f} '.format(avg_rmse))
```

**alpha 0,0.1,1,10,100   RMSE **

```python
# Ridge  alpha   
alphas = [0 , 0.1 , 1 , 10 , 100]

# alphas list  iteration alpha   rmse .
for alpha in alphas :
    ridge = Ridge(alpha = alpha)
    
    #cross_val_score  5 fold  RMSE 
    neg_mse_scores = cross_val_score(ridge, X_data, y_target, scoring="neg_mean_squared_error", cv = 5)
    avg_rmse = np.mean(np.sqrt(-1 * neg_mse_scores))
    print('alpha {0}   5 folds   RMSE : {1:.3f} '.format(alpha,avg_rmse))
```

** alpha     , alpha   plt.subplots   **

```python
import matplotlib.pyplot as plt
#  alpha       5     
fig, axs = plt.subplots(figsize=(18, 6), nrows=1, ncols=5)
#  alpha        DataFrame 
coeff_df = pd.DataFrame()

# alphas           . pos axis  
for pos, alpha in enumerate(alphas):
    ridge = Ridge(alpha=alpha)
    ridge.fit(X_data, y_target)
    
    # alpha     Series   DataFrame  .
    coeff = pd.Series(data=ridge.coef_, index=X_data.columns)
    colname = 'alpha:'+str(alpha)
    coeff_df[colname] = coeff
    
    #    alpha    .     
    coeff = coeff.sort_values(ascending=False)
    axs[pos].set_title(colname)
    axs[pos].set_xlim(-3, 6)
    sns.barplot(x=coeff.values, y=coeff.index, ax=axs[pos])

# for    show   alpha     DataFrame 
plt.show()
```
